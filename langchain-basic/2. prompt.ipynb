```python
from langchain_ollama import ChatOllama

llm = ChatOllama(model="llama3.2:1b")

# llm.invoke(0) # 0을 입력으로 받아서 모델이 어떻게 반응하는지 확인하는 코드 -> 허용되지 않는 문법
```

```python
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate(
    template="What is the capital of {country}?", # placeholder를 이용하여 템플릿을 정의
    input_variables=["country"],
)

prompt = prompt_template.invoke({"country": "France"})

print(prompt)

llm.invoke(prompt)
```

```python
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage

message_list = [
    SystemMessage(content="You are a helpful assistant."),
    AIMessage(content="What can I help you with?"),
    HumanMessage(content="What is the capital of France?")] # France 대신에 {country}로 작성하면 바로 변수가 invoke에 전달되지 않음

llm.invoke(message_list)
```

```python
# 프롬프트에 invoke 메서드를 사용하여 메시지 리스트를 전달하는 코드 -> 허용되지 않는 문법
# 이를 위해서는 from_messages 함수를 사용하여 메시지 리스트를 프롬프트로 변환해야 함

from langchain_core.prompts import ChatPromptTemplate

chat_prompt_template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant."),
        ("ai", "What can I help you with?"),
        ("human", "What is the capital of France?")
    ]
)

chat_promt = chat_prompt_template.invoke({"country": "France"})

print(chat_promt)
```
